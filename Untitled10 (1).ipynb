```python
# prompt: convert these columns gasLimit  , gasPrice , blockNumber	,tokenIn/lastPriceUSD	tokenIn/lastPriceBlockNumber, tokenOut/lastPriceBlockNumber	tokenOut/lastPriceUSD	 to int or numeric so that i can consider it in my graphs

import pandas as pd
df = pd.read_csv("/content/DeFi final.csv")

# Define the columns to convert
columns_to_convert = ['gasLimit', 'gasPrice', 'blockNumber',
                      'tokenIn/lastPriceUSD', 'tokenIn/lastPriceBlockNumber',
                      'tokenOut/lastPriceBlockNumber', 'tokenOut/lastPriceUSD']

# Convert columns to numeric, handling errors
for col in columns_to_convert:
    try:
      df[col] = pd.to_numeric(df[col], errors='coerce')
    except KeyError as e:
      print(f"Warning: Column '{e}' not found in DataFrame. Skipping conversion.")


df.info()
```

```python
# prompt: can you convert float64 to int

# Convert the specified columns to integers, handling potential errors
for col in columns_to_convert:
    try:
        # Attempt to convert to integer, coercing errors to NaN
        df[col] = df[col].astype(int, errors='ignore')

        # Fill NaN values resulting from conversion errors with 0 (or another appropriate value)
        df[col] = df[col].fillna(0).astype(int)
    except KeyError as e:
        print(f"Warning: Column '{e}' not found in DataFrame. Skipping conversion.")
    except Exception as e:
        print(f"Error converting column '{col}': {e}")
        # Handle other potential errors like non-numeric data


df.info()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# Load your dataframe
# df = pd.read_csv("your_data.csv")  # Replace with your actual file path

# --- Step 1: Select relevant numeric columns ---
numeric_cols = [
    "amountIn", "amountInUSD", "amountOut", "amountOutUSD",
    "gasLimit", "gasPrice", "gasUsed", "blockNumber",
    "tokenIn/lastPriceUSD", "tokenIn/lastPriceBlockNumber",
    "tokenOut/lastPriceUSD", "tokenOut/lastPriceBlockNumber"
]

# --- Step 2: Convert object columns to numeric ---
# Coerce errors to NaN so we can handle them
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# --- Step 3: Handle missing values ---
# You can either fill them or drop them. Let's fill with median here:
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# --- Step 4: Feature Scaling ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[numeric_cols])

# --- Step 5: Train-test split (optional for unsupervised) ---
X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)

# --- Step 6: Fit Isolation Forest ---
iso_forest = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)
iso_forest.fit(X_train)

# --- Step 7: Predict Anomaly Scores and Labels ---
# -1 = anomaly, 1 = normal
anomaly_labels = iso_forest.predict(X_test)
anomaly_scores = iso_forest.decision_function(X_test)

# You can also attach them to the test data:
df_test = pd.DataFrame(X_test, columns=numeric_cols)
df_test["anomaly_label"] = anomaly_labels
df_test["anomaly_score"] = anomaly_scores

# View some of the anomalies
anomalies = df_test[df_test["anomaly_label"] == -1]
print(anomalies.head())

print(anomalies)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df_test["label"] = df_test["anomaly_label"].map({1: "Normal", -1: "Anomaly"})
sns.scatterplot(data=df_test, x="amountInUSD", y="gasPrice", hue="label")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_test)
df_test["PCA1"], df_test["PCA2"] = X_pca[:, 0], X_pca[:, 1]

sns.scatterplot(data=df_test, x="PCA1", y="PCA2", hue="label")
plt.show()
```
